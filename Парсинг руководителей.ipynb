# Парсер руководителей

Используемый ресурс: https://uchet.kz/

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
df = pd.read_excel("Лист Microsoft Excel — исходник.xlsx")
```

```python
#если бины даны не в формате 12 цифр, то можно добавить нули перед бином таким способом
id_list = df['БИН'].astype(str).str.zfill(12)
```

```python
id_list
```

```python
data = []  # Список для хранения данных
```

```python
for id in id_list:
    # Формирование URL с использованием переменной id
    url = f'https://pk.uchet.kz/search/bin/{id}'
    
    # Отправка HTTP-запроса GET на сервер
    response = requests.get(url)
    
    # Проверка статуса ответа сервера
    if response.status_code == 200:
        # Парсинг HTML-контента ответа с использованием BeautifulSoup
        soup = BeautifulSoup(response.text, 'html.parser')
        
        # Поиск всех элементов с классом 'info-block__object-block', которые содержат нужные данные
        headings = soup.find_all('div', class_='info-block__object-block')

        # Перебор найденных элементов и вывод их текстового содержимого
        for heading in headings:
            # Удаление пробелов и переносов строк в начале и в конце строки
            print(heading.text.strip())
            
         # Добавление в список данных
        data.append(heading.text)
        
    else:
        # Вывод сообщения об ошибке, если ответ сервера не равен 200 OK
        print(f"Ошибка при запросе страницы {url}. Код ответа: {response.status_code}")
        data.append(f"-")
```

```python
data
```

```python
# Чистка данных
cleaned_names = [name.strip() for name in data]
```

```python
cleaned_names
```

```python
# Добавление данных в датафрейм Руководитель, то есть добавление данных в столбец Руководитель
df['Руководитель'] = cleaned_names
```
